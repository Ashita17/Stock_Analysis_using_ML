```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = pd.read_csv('JIOFIN.NS.csv')
jiofin_close = df['Close']
jiofin_close
```

```python
jiofin_close.reset_index(drop=True,inplace=True)
jiofin_close.shape
```

```python
jiofin_close = pd.DataFrame(jiofin_close)
jiofin_close.plot();
```

```python
train = jiofin_close.iloc[:30]
test = jiofin_close.iloc[30:]
test.size
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 17 
batch_size = 23

generator = TimeseriesGenerator(scaled_train,scaled_train,length=length,batch_size=batch_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN ,Dropout , LSTM

```

```python
n_features = 1
```

```python
model = Sequential()

model.add(SimpleRNN(23,input_shape=(length,n_features)))
model.add(Dropout(rate=0.2))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator,epochs=13,shuffle=False)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot();
```

```python
first_eval_batch = scaled_train[-length:]
first_eval_batch = first_eval_batch.reshape((1,length,1))
first_eval_batch
```

```python
model.predict(first_eval_batch)
```

```python
test_prediction = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(test)):

    current_pred = model.predict(current_batch)[0]
    test_prediction.append(current_pred) 
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_prediction = scaler.inverse_transform(test_prediction)
```

```python
test['prediction'] = true_prediction
test
```

```python
test.plot(figsize=(12,8));
```

LSTM Model

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience= 2)
```

```python
leng = length -1
generator = TimeseriesGenerator(scaled_train,scaled_train,length=leng,batch_size=25)
validation_generator = TimeseriesGenerator(scaled_test,scaled_test,length=leng,batch_size=25)
```

```python
model = Sequential()

model.add(LSTM(30,input_shape=(leng,n_features)))
model.add(Dropout(rate=0.2))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit_generator(generator,epochs=20,validation_data=validation_generator,callbacks=[early_stop],shuffle=False)
```

```python
lossess = pd.DataFrame(model.history.history)
lossess.plot();
```

```python
test_prediction = []
first_eval_batch = scaled_train[-leng:]
current_batch = first_eval_batch.reshape((1,leng,n_features))

for i in range(len(test)):

    current_pred = model.predict(current_batch)[0]
    test_prediction.append(current_pred) 
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_prediction = scaler.inverse_transform(test_prediction)
test['prediction_LSTM'] = true_prediction
test.plot(figsize=(12,8));
```
